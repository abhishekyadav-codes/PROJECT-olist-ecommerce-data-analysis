```python
# Q11. Monthly revenue trend nikalo.
month_trend=df.merge(df_order,on='order_id')
month_trend
```

```python
month_trend['order_purchase_timestamp'] = pd.to_datetime(month_trend['order_purchase_timestamp'])
```

```python
month_trend['month'] = month_trend['order_purchase_timestamp'].dt.month_name()
month_trend
```

```python
a=month_trend.groupby('month')['revenue'].sum()
a
```

```python
plt.figure(figsize=(20,10))
a.plot(kind='line',marker='o')
plt.xlabel('MONTHS')
plt.ylabel('REVENUE')
```

```python
# Q12. Kis month me sabse zyada orders aaye?
a=month_trend.groupby('month')['order_item_id'].sum()
a.sort_values(ascending=False)
```

```python
# Q13. Year wise growth rate calculate karo.
month_trend['year']=month_trend['order_purchase_timestamp'].dt.year
```

```python
a=month_trend.groupby('year')['revenue'].sum()
a
```

```python
plt.figure(figsize=(10,5))
plt.bar(a.index,a.values)
plt.xlabel('YEAR')
plt.ylabel('REVENUE')
plt.show()
```
